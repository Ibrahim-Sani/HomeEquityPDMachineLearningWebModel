```python

import pandas as pd


df = pd.read_csv('SECONDARY_FACILITY_ISS_TOOL_results.csv')

print(df.shape)
```

```python

newDF = df["grp1:group1:selected_facility"] + df["grp2:ggp2:selected_facility1"]

df["Facilities_Combined"]= df["grp1:group1:selected_facility"].astype(str) + df["grp2:ggp2:selected_facility1"].astype(str)
df["Facilities_Combined"]= df["Facilities_Combined"].str.replace("nan","")

df.sort_values('are_you_the_team_lead', ascending=False)

df.groupby(["Facilities_Combined"]).first().reset_index()

NewDF = df.groupby(["Facilities_Combined"]).first().reset_index()

NewDF.to_csv('SHFISSmergedMay4.csv')
```
